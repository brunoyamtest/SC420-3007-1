```python
#Домашнее задание
#Easy
#Изучите распределение размеров компаний (company_size) с помощью метода .value_counts(). Опишите своими словами, что вы увидели.
# Ваш код ниже:
import pandas as pd
df = pd.read_csv('ds_salaries_mod.csv')
df['company_size'].value_counts()
```

```python
#Normal
#Сделайте выборку за 2023 год. Постройте по ней сводную таблицу, сгруппированную по размеру компании-работодателя (company_size),
#со средней и медианной зарплатой в долларах. Не забывайте также указать количество.
# Ваш код ниже:
df.query('work_year == 2023').groupby('company_size')['salary_in_usd'].agg(['count', 'mean', 'median'])
```

```python
# Аналогично исследуйте 2022 год. Есть ли отличия?
# Ваш код ниже:
df.query('work_year == 2022').groupby('company_size')['salary_in_usd'].agg(['count', 'mean', 'median'])
```

```python
# Ваш код:
# При анализе было выявлено что начения в столбцах 'salary' и 'salary_in_usd' когда salary_currency == USD совпаддают,
# по этому было принято решение недостаюие начения в 'salary_in_usd' дополнить и 'salary' 
df_clean = df.copy()
df_clean['salary_in_usd'] = df_clean['salary_in_usd'].fillna(df_clean['salary'])
df_clean[df['salary_in_usd'].isna()]
```

```python
#Дополнительно: попробуйте свести данные в столбце job_title к 4 основным профессиям
#(Data Scientist, Data Analyst, Data Engineer, ML Engineer), где это возможно.
# Ваш код:
options = ['Data Scientis', 'Data Analyst', 'Data Engineer', 'Machine Learning Engineer']     
rslt_df = df_clean[df_clean['job_title'].isin(options)]     
rslt_df
```

```python
#Сделайте выборку за 2023 год. Постройте по ней сводную таблицу, сгруппированную по профессии (job_title), со средней и медианной зарплатой в долларах. Не забывайте также указать количество. (Если у вас не получилось привести в порядок столбец job_title,
#выведите только 4 самых распространенных профессии.)
# Ваш код:
df_clean.query('work_year == 2023').groupby('job_title')['salary_in_usd'].agg(['count', 'mean', 'median']
    .sort_values(by='count', ascending=False).head(4))
```
